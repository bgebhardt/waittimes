```python
# from [How do I read multiple CSV files from multiple directories using Python? - Quora](https://www.quora.com/How-do-I-read-multiple-CSV-files-from-multiple-directories-using-Python)
df_list = []

for file in os.listdir('.'):
    df = pd.read_csv(file)
    df_list.append(df)
final_df = pd.concat(df_list)
```
